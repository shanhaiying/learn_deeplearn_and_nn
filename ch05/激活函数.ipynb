```python
import tensorflow as tf
# 二维张量
t=tf.constant([[1,3] ,[2,0]], tf.float32)
#sigmoid激活
result=tf.nn.sigmoid(t)
#创建会话
session=tf.Session()
#打印结果
print(session.run(result))
```

# RELU 激活函数

```python
import tensorflow as tf
# 变量
x=tf.Variable(tf.constant([[2,1,3]] ,tf.float32))
w=tf.constant([[2],[3], [4]] ,tf.float32)
#函数g
g=tf.matmul(x,w)
#函数f=relu(g)
f=tf.nn.relu(g)
#计算f在(2,1,3)处的导数
gradient=tf.gradients(f,[g,x])
session=tf.Session()
session.run(tf.global_variables_initializer())
#打印结果
print(session.run(gradient))
```

# Leaky_relu 激活函数

```python
import tensorflow as tf
x=tf.Variable(tf.constant([[2,1,3]] ,tf.float32))
w=tf.constant([[2],[3],[4]] ,tf.float32)
#函数g
g=tf.matmul(x,w)
# 激活函数
f=tf.nn.leaky_relu(g,alpha=0.2)
#牛顿梯度下降法
opti=tf.train.GradientDescentOptimizer(0.5).minimize(f)
session=tf.Session()
session.run(tf.global_variables_initializer())
#打印结果
for i in range(3):
    session.run(opti)
    print('第%d次迭代的值'%(i+1))
    print(session.run(x))
```

# crelu激活函数

```python
import tensorflow as tf
t=tf.constant([-2,0,1], tf.float32)
# crelu激活函数
r=tf.nn.crelu(t)
# 创建会话
session= tf.Session()
#打印结果
print(session.run(r))
```

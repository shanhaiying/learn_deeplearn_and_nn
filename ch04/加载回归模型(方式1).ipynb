```python
import tensorflow as tf
#初始化两个变量, 变量形状与model.cpkt文件中的必须相等
v1=tf.Variable(tf.constant([11,12,13],tf.float32), dtype=tf.float32, name='v1')
v2=tf.Variable(tf.constant([14,15],tf.float32), dtype=tf.float32, name='v2')

#声明一个tf.train.Saver对象
saver = tf.train.Saver()

with tf.Session() as sess:
    # 加载model.ckpt文件
    saver.restore(sess, './model.ckpt')
    #打印两个变量的值
    print(sess.run(v1))
    print(sess.run(v2))
    
session.close()
```

```python
import numpy as np
f=np.array(
[
    [10,2,8],
    [5,12,3]
],np.complex64)
#第一步,对每一列进行傅立叶变换
f_0_fft=np.fft.fft(f,axis=0)
print('第一步,对每一列进行傅立叶变换')
print(f_0_fft)
# 第二步, 针对第1步得到的结果, 分别对每一行进行傅立叶变换
f_0_1_fft=np.fft.fft(f_0_fft,axis=1)
print('第二步, 针对第1步得到的结果, 分别对每一行进行傅立叶变换')
print(f_0_1_fft)
```

```python
#第一步,对每一行进行傅立叶变换
f_1_fft=np.fft.fft(f,axis=1)
print('第一步,对每一行进行傅立叶变换')
print(f_1_fft)
# 第二步, 针对第1步得到的结果, 分别对每一列进行傅立叶变换
f_1_0_fft=np.fft.fft(f_1_fft,axis=0)
print('第二步, 针对第1步得到的结果, 分别对每一列进行傅立叶变换')
print(f_1_0_fft)
```

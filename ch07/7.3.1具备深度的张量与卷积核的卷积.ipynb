```python
import tensorflow as tf
# 1个长度为3, 深度为3的张量
x=tf.constant(
[
    [[2,5,2] , [6,1,-1], [7,9,-5]]
], tf.float32)

# 1个长度为2, 深度为3的卷积核
k=tf.constant(
[
    [[-1],[5],[4]] , 
    [[2],[1],[6]]
]
    ,tf.float32)

# 一维same卷积
v_conv1d_k= tf.nn.conv1d(x,k,1,'SAME')
#创建会话
session=tf.Session()
print(session.run(v_conv1d_k))
```

```python
import tensorflow as tf
#输入形状为[1,4,4,1]的张量
value2d=tf.constant([
    [
        [[2],[3],[8],[-2]],
        [[6],[1],[5],[9]],
        [[7],[2],[-1],[8]],
        [[1],[4],[3],[5]],
    ]
        
],tf.float32)

#最大值池化操作
value2d_maxPool=tf.nn.max_pool(value2d,(1,2,3,1),[1,1,1,1],'SAME')
session=tf.Session()
print(session.run(value2d_maxPool))
```

```python
import tensorflow as tf
#输入形状为[1,4,4,1]的张量
value3d=tf.constant([
    #第一个3行3列2深度的三维张量
    [
        [[2,5],[3,3],[8,2]],
        [[6,1],[1,2],[5,4]],
        [[7,9],[2,-3],[-1,3]]
    ]  
],tf.float32)

#平均值池化操作
value2d_maxPool=tf.nn.avg_pool(value3d,(1,2,2,1),[1,2,2,1],'SAME')
session=tf.Session()
print(session.run(value2d_maxPool))
```

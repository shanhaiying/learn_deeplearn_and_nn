```python
import tensorflow as tf
```

```python
#二维张量
value2d=tf.constant(
[
    [5,1,4,2],
    [3,9,5,7]
],
    tf.float32
)
#求沿1方向上, 即每一行的最大值的位置索引
result=tf.argmax(value2d, axis=1)
session=tf.Session()
#打印结果
print('求沿1方向上, 即每一行的最大值的位置索引:',session.run(result))

#求沿0方向上, 即每一列的最大值的位置索引
result=tf.argmax(value2d, axis=0)
print('求沿0方向上, 即每一列的最大值的位置索引',session.run(result))
```
